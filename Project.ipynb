```python
from google.colab import files
import pandas as pd

# Step 1: Upload the file
uploaded = files.upload()

# Step 2: Load the file into a DataFrame
micro_ken = pd.read_csv('micro_ken.csv')

# Step 3: Check the data
print(micro_ken.info())
print(micro_ken.head())
```

```python
# Step 4: Define the columns to keep
columns_to_keep = [
    'economy', 'age', 'female', 'educ', 'inc_q', 'emp_in', 'urbanicity_f2f',
    'account', 'account_fin', 'account_mob', 'receive_transfers', 'mobileowner',
    'internetaccess', 'anydigpayment', 'merchantpay_dig'
]

# Step 5: Check which columns are missing
missing_columns = [col for col in columns_to_keep if col not in micro_ken.columns]
print("Missing columns:", missing_columns)

# Step 6: Select the relevant columns
cleaned_data = micro_ken[columns_to_keep]
print(cleaned_data.info())
```

```python
# Rename columns for better clarity
renamed_columns = {
    'economy': 'Country',
    'age': 'Age',
    'female': 'Gender',
    'educ': 'Education_Level',
    'inc_q': 'Income_Quintile',
    'emp_in': 'Employment_Status',
    'urbanicity_f2f': 'Urbanicity',
    'account': 'Has_Account',
    'account_fin': 'Has_Financial_Account',
    'account_mob': 'Has_Mobile_Account',
    'receive_transfers': 'Received_Transfers',
    'mobileowner': 'Owns_Mobile',
    'internetaccess': 'Has_Internet_Access',
    'anydigpayment': 'Made_Digital_Payment',
    'merchantpay_dig': 'Paid_Merchant_Digitally'
}

# Apply renaming
cleaned_data = cleaned_data.rename(columns=renamed_columns)
print(cleaned_data.info())
```

```python
# Check for missing values
print(cleaned_data.isnull().sum())

# Drop rows with missing values
cleaned_data_dropped = cleaned_data.dropna()
print("Rows before dropping missing values:", len(cleaned_data))
print("Rows after dropping missing values:", len(cleaned_data_dropped))
```

```python
# Save the cleaned data to a CSV
cleaned_data_dropped.to_csv('final_cleaned_micro_kenya.csv', index=False)

# Download the file
from google.colab import files
files.download('final_cleaned_micro_kenya.csv')
```

```python
import altair as alt
import pandas as pd

# Step 1: Upload the file
uploaded = files.upload()

# Step 2: Load the file into a DataFrame
micro_ken = pd.read_csv('final_cleaned_micro_kenya.csv')

# Ensure data types are correct
# Use 'micro_ken' instead of 'data' to access the correct DataFrame
micro_ken['Owns_Mobile'] = micro_ken['Owns_Mobile'].astype(int)
micro_ken['Has_Internet_Access'] = micro_ken['Has_Internet_Access'].astype(int)
micro_ken['Has_Financial_Account'] = micro_ken['Has_Financial_Account'].astype(int)
micro_ken['Made_Digital_Payment'] = micro_ken['Made_Digital_Payment'].astype(int)
micro_ken['Urbanicity'] = micro_ken['Urbanicity'].astype(int)

# Prepare data for Altair visualizations

# Visualization 1: Bar Chart - Digital Payment Usage by Mobile Ownership
# Use 'micro_ken' instead of 'data' in the chart definitions
chart1 = alt.Chart(micro_ken).mark_bar().encode(
    x=alt.X('Owns_Mobile:N', title='Owns Mobile (1=Yes, 2=No)'),
    y=alt.Y('mean(Made_Digital_Payment):Q', title='Mean Digital Payment Usage'),
    color=alt.Color('Owns_Mobile:N', legend=None)
).properties(
    title='Digital Payment Usage by Mobile Ownership',
    width=400,
    height=300
)

# Visualization 2: Scatter Plot - Mobile Ownership vs Mobile Money Account
chart2 = alt.Chart(micro_ken).mark_point().encode(
    x=alt.X('Owns_Mobile:N', title='Owns Mobile (1=Yes, 2=No)'),
    y=alt.Y('Has_Mobile_Account:N', title='Has Mobile Money Account (1=Yes, 0=No)'),
    color=alt.Color('Owns_Mobile:N', legend=None),
    tooltip=['Owns_Mobile', 'Has_Mobile_Account']
).properties(
    title='Mobile Ownership vs Mobile Money Account',
    width=400,
    height=300
)

# Display the charts side-by-side for comparison
chart1 | chart2
```

Does mobile phone ownership correlate with access to digital payments in Kenya?

These charts strongly support the hypothesis that technological advancements like mobile phone ownership are critical enablers of financial inclusion in Kenya. Mobile phones not only facilitate financial transactions but also increase access to mobile money accounts, a key component of digital financial services.

```python
import altair as alt
import pandas as pd

# Step 1: Upload the file
uploaded = files.upload()

# Step 2: Load the file into a DataFrame
micro_ken = pd.read_csv('final_cleaned_micro_kenya.csv')

# Visualization 5: Grouped Bar Chart - Digital Payment Usage by Urbanicity
chart5 = alt.Chart(micro_ken).mark_bar().encode(
    x=alt.X('Urbanicity:N', title='Urbanicity (1=Urban, 2=Rural)'),
    y=alt.Y('mean(Made_Digital_Payment):Q', title='Mean Digital Payment Usage'),
    color=alt.Color('Urbanicity:N', legend=None)
).properties(
    title='Digital Payment Usage by Urbanicity',
    width=400,
    height=300
)

# Visualization 6: Stacked Bar Chart - Financial Inclusion Metrics by Urbanicity
urban_data = micro_ken.groupby('Urbanicity').agg(
    Financial_Account=('Has_Financial_Account', 'mean'),
    Digital_Payment=('Made_Digital_Payment', 'mean')
).reset_index()

chart6 = alt.Chart(urban_data).transform_fold(
    ['Financial_Account', 'Digital_Payment'],
    as_=['Metric', 'Value']
).mark_bar().encode(
    x=alt.X('Urbanicity:N', title='Urbanicity (1=Urban, 2=Rural)'),
    y=alt.Y('Value:Q', title='Percentage'),
    color=alt.Color('Metric:N', title='Metric'),
    tooltip=[alt.Tooltip('Urbanicity:N', title='Urbanicity'),
             alt.Tooltip('Metric:N', title='Metric'),
             alt.Tooltip('Value:Q', title='Percentage')]
).properties(
    title='Financial Inclusion Metrics by Urbanicity',
    width=400,
    height=300
)

# Display the charts side-by-side
chart5 | chart6
```

What role does urbanicity play in the adoption of digital payments and financial inclusion in Kenya?

Urban residents have higher digital payment usage and account ownership than rural residents, highlighting disparities in financial inclusion and access to financial technology between urban and rural areas.

```python
import altair as alt
import pandas as pd

# Step 1: Upload the file
uploaded = files.upload()

# Step 2: Load the file into a DataFrame
micro_ken = pd.read_csv('final_cleaned_micro_kenya.csv')

# Replace column names with verified names
correlation_data = micro_ken[[
    'Owns_Mobile', 'Has_Internet_Access', 'Has_Financial_Account', 'Made_Digital_Payment'
]]

# Compute correlation matrix
correlation_matrix = correlation_data.corr().reset_index().melt('index')
correlation_matrix.columns = ['Variable 1', 'Variable 2', 'Correlation']

# Create an interactive heatmap
interactive_heatmap = alt.Chart(correlation_matrix).mark_rect().encode(
    x=alt.X('Variable 1:N', title=''),
    y=alt.Y('Variable 2:N', title=''),
    color=alt.Color('Correlation:Q', scale=alt.Scale(scheme='viridis')),
    tooltip=['Variable 1', 'Variable 2', 'Correlation']
).properties(
    title='Interactive Correlation Heatmap: Technology and Financial Inclusion',
    width=400,
    height=400
).interactive()

interactive_heatmap
```

This correlation heatmap shows the relationships between mobile ownership, internet access, financial accounts, and digital payments in Kenya. Mobile ownership and internet access positively influence financial inclusion metrics.

```python
import pandas as pd
import altair as alt

# Step 1: Upload the file
uploaded = files.upload()

# Step 2: Load the file into a DataFrame
mobile_data = pd.read_csv("mobile_data_kenya.csv")
internet_users = pd.read_csv("internet_users_kenya.csv")
mobile_money_accounts = pd.read_csv("mobile_money_accounts_kenya.csv")

# Add a column to distinguish indicators
mobile_data["Indicator"] = "Mobile Supscriptions (Millions)"
internet_users["Indicator"] = "Internet Users (% of Population)"
mobile_money_accounts["Indicator"] = "Mobile Money Accounts (% of Adults)"

# Combine datasets
all_data = pd.concat([mobile_data, internet_users, mobile_money_accounts], ignore_index=True)

# Ensure the 'date' and 'value' columns are numeric
all_data["date"] = pd.to_numeric(all_data["date"], errors="coerce")
all_data["value"] = pd.to_numeric(all_data["value"], errors="coerce")
all_data = all_data.dropna()  # Drop rows with missing data

# Create an indicator dropdown selection
dropdown = alt.binding_select(
    options=list(all_data["Indicator"].unique()),  # Get unique indicator names
    name="Select Indicator: "
)
indicator_selection = alt.selection_point(fields=["Indicator"], bind=dropdown)

# Create the interactive line chart
chart = alt.Chart(all_data).transform_filter(
    indicator_selection
).mark_line(point=True).encode(
    x=alt.X("date:O", title="Year"),
    y=alt.Y("value:Q", title="Value"),
    color=alt.Color("Indicator:N", legend=None),
    tooltip=[
        alt.Tooltip("date:O", title="Year"),
        alt.Tooltip("value:Q", title="Value"),
        alt.Tooltip("Indicator:N", title="Indicator")
    ]
).add_selection(
    indicator_selection
).properties(
    title="Technological Growth and Digital Financial Inclusion in Kenya",
    width=700,
    height=400
)

# Display the chart
chart
```

 This chart illustrated that these trends support the main question, showing technological growth, via mobile subscriptions, internet access, and mobile money accounts which directly correlates with increased financial inclusion in Kenya over time.
